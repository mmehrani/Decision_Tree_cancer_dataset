```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import DecisionTreeClassifier,plot_tree,export_graphviz
import pandas as pd
```

```python
data = pd.read_csv('wbc.csv',index_col = 'id')
data
```

```python
data[data['diagnosis'] == 'B'].describe()
```

# بررسی رفتار داده‌ها

```python
s1 = data[data['diagnosis'] == 'M'].describe().min()
s2 = data[data['diagnosis'] == 'M'].describe().max()
df_malignant = pd.DataFrame(pd.concat([s1,s2],axis = 1))
df_malignant.rename(columns = {0:'min',1:'Max'})
```

```python
s1 = data[data['diagnosis'] == 'B'].describe().min()
s2 = data[data['diagnosis'] == 'B'].describe().max()
df_malignant = pd.DataFrame(pd.concat([s1,s2],axis = 1))
df_malignant.rename(columns = {0:'min',1:'Max'})
```

# به کارگیری مدل درخت

```python
X = data.drop('diagnosis',axis = 1)
y = data['diagnosis'].map({'M':1,'B':0})
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
tree_model = DecisionTreeClassifier(criterion = 'entropy', max_depth= 4)
tree_model.fit(X_train,y_train)
y_predicted = tree_model.predict(X_test)
```

```python
accuracy_score(y_predicted,y_test)
```
